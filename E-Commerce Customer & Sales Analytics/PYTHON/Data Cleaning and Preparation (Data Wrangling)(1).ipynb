## Fixing Datatypes

```python
# fixing date colunms datatype
customers['signup_date'] = pd.to_datetime(customers['signup_date'])
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
# creating month column
orders['month'] = orders['order_date'].dt.month_name()
```

```python
# creating day column
orders['day'] = orders['order_date'].dt.day_name()
```

## MERGE TABLES

```python
df = orders.merge(
    customers,
    on='customer_id',
    how='left'
)

df = df.merge(
    order_items,
    on='order_id',
    how='left'
)

df = df.merge(
    products,
    on='product_id',
    how='left'
)
```

```python
df
```

```python
df.isnull().sum()
```

```python
# We found some null values after merging the tables i.e some orders without products.
```

```python
# checking which orders are without products
orders_without_items = orders[
    ~orders['order_id'].isin(order_items['order_id'])
]

print(orders_without_items.head())
```

```python
# Remove rows without products
df = df.dropna(subset=['product_id'])
```

```python
df
```

```python
# Checking Null values
df.isnull().sum()
```

```python
# Add Revenue column tp table

df['revenue'] = df['quantity']*df['price']
```
